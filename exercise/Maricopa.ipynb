```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
hdf = pd.read_excel('Maricopa.xlsx')
hdf_2019 = hdf[hdf['year_sold'] == 2019 ]
hdf_2019
```

```python
results = smf.ols('sale_price ~ area + pool_area + garage_count', data=hdf_2019).fit()
print(results.summary())
```

```python
results = smf.ols('sale_price ~ area + pool_area + garage_count + city', data=hdf_2019).fit()
print(results.summary())
```
