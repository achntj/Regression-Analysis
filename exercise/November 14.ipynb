```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel("Maricopa.xlsx")
df
```

```python
results = smf.ols("sale_price ~ area + single_story", data=df).fit()
print(results.summary())
```

```python
random = np.random.uniform(0, 1, size=len(df['sale_price']))
df['random'] = random
```

```python
df
```

```python
df = df.drop(df[(df.random < 0.5) & (df.single_story == 1)].index)
df
```

```python
results2 = smf.ols("sale_price ~ area + single_story", data=df).fit()
print(results2.summary())
```
