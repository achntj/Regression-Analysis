```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel("Maricopa.xlsx")
df
```

```python
results = smf.ols("sale_price ~ area", data=df).fit()
print(results.summary())
```

```python
error = np.random.normal(0, np.std(df['sale_price']), size=len(df['sale_price']))
df['price_with_error'] = df['sale_price'] + error
```

```python
df
```

```python
results2 = smf.ols("price_with_error ~ area", data=df).fit()
print(results2.summary())
```

```python
error = np.random.normal(0, np.std(df['area']), size=len(df['area']))
df['area_with_error'] = df['area'] + error
df
```

```python
results3 = smf.ols("sale_price ~ area_with_error", data=df).fit()
print(results3.summary())
```
